```python
import numpy as np
import os
import copy
import warnings

import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import seaborn as sns
sns.set()
sns.set_style("white")
sns.set_palette(sns.diverging_palette(220, 20, n=6))
```

```python
nOrb = 3

beta = 16
U = 8
initCond = 1

SAVEDIR = "../plots/MeanFieldTMDnanoribbon/"
if not os.path.exists(SAVEDIR):
    os.makedirs(SAVEDIR)
```

## Comparisons

```python
plt.ylabel(r'$\left\langle m_y \right\rangle$')
plt.xlabel(r'$y$')
colors = ['r', 'b', 'g', 'orange']

for idy, ny in enumerate([8, 16, 32]):
    for idx, nx in enumerate([128, 256, 512, 1024]):
        SAVESUBDIR = SAVEDIR + "Nx=" + str(nx) +\
            "_Ny=" + str(ny) + "_U=" + str(U) + "_beta=" + str(beta) + "/"
        nUp = np.loadtxt(SAVESUBDIR + "nUp.txt")
        nDown = np.loadtxt(SAVESUBDIR + "nDown.txt")
        sns.set_style("white")
        nUpSite = np.zeros(ny)
        nDownSite = np.zeros(ny)
        for i in range(ny):    
            nUpSite[i] = nUp[nOrb * i] + nUp[nOrb * i + 1] + nUp[nOrb * i + 2]
            nDownSite[i] = nDown[nOrb * i] + nDown[nOrb * i + 1] + nDown[nOrb * i + 2]
        
        plt.figure(idy)
        plt.plot((nUpSite-nDownSite), markersize=4, linewidth = 0.5 , c = colors[idx], marker = 'o',\
                label = r'$Nx=$' + str(nx))
        plt.title(r'$Ny=$' + str(ny))
    plt.savefig(SAVEDIR + "comparison_Nx_" +\
        "_Ny=" + str(ny) + "_U=" + str(U) + "_beta=" + str(beta)\
        + ".png", dpi = 500, bbox_inches='tight')
    
```

```python
sns.set_palette("Paired")
sns.set_style("white")

plt.ylabel(r'$\left\langle m_y \right\rangle$')
plt.xlabel(r'$y$')

for idy, ny in enumerate([8, 16, 32, 64, 128]):
    for idx, nx in enumerate([512]):
        SAVESUBDIR = SAVEDIR + "/Nx=" + str(nx) +\
            "_Ny=" + str(ny) + "_U=" + str(U) + "_beta=" + str(beta) + "/"
        nUp = np.loadtxt(SAVESUBDIR + "nUp.txt")
        nDown = np.loadtxt(SAVESUBDIR + "nDown.txt")
        sns.set_style("white")
        nUpSite = np.zeros(ny)
        nDownSite = np.zeros(ny)
        for i in range(ny):    
            nUpSite[i] = nUp[nOrb * i] + nUp[nOrb * i + 1] + nUp[nOrb * i + 2]
            nDownSite[i] = nDown[nOrb * i] + nDown[nOrb * i + 1] + nDown[nOrb * i + 2]
        
        plt.plot((nUpSite-nDownSite), markersize=2.5, linewidth = 0.5 , marker = 'o',
                label = r'$N_y=$' + str(ny))
        plt.legend(bbox_to_anchor=(1, 1),)
        plt.title(r'$N_k=$' + str(nx))
    plt.savefig(SAVEDIR + "/comparison_Ny" + ".png", dpi = 500, bbox_inches='tight')
    
```
