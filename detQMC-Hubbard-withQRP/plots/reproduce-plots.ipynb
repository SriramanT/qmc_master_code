```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
cwd = os.getcwd()
import seaborn as sns

sns.set()
sns.set_palette("Blues_r")

# Are you testing (0) or making plots (1)?

SAVEPLOTS = bool(1) # 0 for false, 1 for true

# To reproduce a plot that was previously made, uncomment the lines below and enter the
# simulation parameters

NSITES = 40
dt = 0.125
beta = 15.
L = 120
t = 1.
U = 4.
mu = 0.
totalMCSweeps = 100
freq = 2
intsize = 60

mainDir = (str(NSITES) + \
             'sites_L=' + str(L) + \
             '_beta=' + str(beta) + \
             '_dt_' + str(dt) + '_t_' + \
             str(t) + '_U_'+ str(U) + '_mu_' + str(mu))

data_to_reproduce = (mainDir + '/data-to-reproduce')

# Load weights to plot

weights = np.loadtxt(data_to_reproduce + '/weights_' + \
                          'totalMCSweeps_' + str(totalMCSweeps) + \
                          '_freq_' + str(freq) + '_intsize_' + str(intsize) + '.txt')


M = weights.size

time = np.arange(M) + 1   #measured in sweeps

sns.set_style("white")
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
plt.scatter(time / L, weights, s = 0.3)
plt.xlabel(r"Space-time sweep")
plt.ylabel(r'$\log \frac{| P(\mathbf{h}) | }{ | P(\mathbf{h_0}) | } $')
if SAVEPLOTS == 1:
    plt.savefig(mainDir + '/weightsVsMCsweep' + str(totalMCSweeps) + 'sweeps' + '_freq_' + str(freq)
            + '_intsize_' + str(intsize) + '.png', dpi = 600)
```
